```python
# prompt: draw a histogram of quantity and pizza name

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.barplot(x='pizza_name', y='quantity', data=df.groupby('pizza_name')['quantity'].sum().reset_index().sort_values(by='quantity', ascending=False))
plt.xticks(rotation=90)
plt.title('Total Quantity Ordered per Pizza Name')
plt.xlabel('Pizza Name')
plt.ylabel('Total Quantity Ordered')
plt.tight_layout()
plt.show()
```

```python
# prompt: make a count graph for pizza category

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="pizza_category")
plt.title('Count of Pizzas by Category')
plt.xlabel('Pizza Category')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
