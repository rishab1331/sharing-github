```python
import pandas as pd
```

```python
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data' , header = None,
                   names = ['sl', 'sw','pl','pw','species'])
```

```python
df = iris.copy()
```

```python
df1 = df['species'].value_counts()
print(df1.values[0], df1.values[1],df1.values[2])
```

```python
df[df['species']=='Iris-virginica']
df[(df.pl > 1.5)]
```

```python
df.groupby('species').agg(['min' , 'max'  , 'mean']).round(2)
```
